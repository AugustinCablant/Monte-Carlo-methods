```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n = 100000
T = int(np.random.uniform(1,100))
l = int(np.random.uniform(1,20))
Z = np.random.poisson(l * T)
X = np.random.uniform(1, T, size = n)
Y = [np.sum((X[0:Z] <= t).astype(int)) for t in range(T)]
plt.figure(figsize=(12,8))
plt.scatter(x = range(len(Y)), y = Y)
plt.show()
```
