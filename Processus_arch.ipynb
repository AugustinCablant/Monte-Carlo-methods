# Processus ARCH

Soit $(\eta _k)$ une suite de variables i.i.d gaussiennes centrées réduites et pour pour tout $k$, $ \mathcal{F} _k = \sigma ( \eta _1, ..., \eta _k)$. 

On définit par récurrence $Z_0 = 1, \space Z_{k+1} = \eta _{k+1} \sqrt{ \alpha _0 + \alpha _1 Z_k ^2}$.

En posant $ M_n = \sum _{k=1} ^{n} Z_k$, on vérifie très facilement que $M_n$ est une martingale.

```python
n = 10000
a0 = 0.5
a1 = 0.7
E = np.random.normal(0,1,size=n)
Z = [1]
for i in range(1,n):
    z2 = Z[i-1] ** 2
    Z.append(E[i] * np.sqrt(a0 + a1 * z2))
M = np.cumsum(Z)

plt.figure(figsize=(12,8))
plt.plot(list(range(len(Z))), Z, label = f'Exemple de réalisation avec n = {n}, alpha_0 = {a0}, alpha_1 = {a1}')
plt.xlabel('t')
plt.ylabel('Z_t')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(list(range(len(M))), M, label = f'Exemple de réalisation avec n = {n}, alpha_0 = {a0}, alpha_1 = {a1}')
plt.xlabel('t')
plt.ylabel('M_t')
plt.legend()
plt.show()
```

```python
l = 12
nu = 0.01
Y = np.exp(l - nu * M)
plt.figure(figsize=(12,8))
plt.plot(list(range(len(Y))), Y, label = f'Exemple de réalisation avec n = {n}, alpha_0 = {a0}, alpha_1 = {a1}, lambda = {l}, nu = {nu}')
plt.xlabel('t')
plt.ylabel('M_t')
plt.legend()
plt.show()
```
